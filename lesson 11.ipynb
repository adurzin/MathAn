### 1. Исследовать ряд на сходимость, используя признак д'Аламбера:
### $\sum_{n=1}^{\infty} \frac{n^{n}}{n!^{2}}
$

### $ q = \lim\limits_{n \to \infty} \frac{(n+1)^{n+1}}{(n+1)!^{2}} \cdot \frac{n!^{2}}{n^{n}} = \lim\limits_{n \to \infty} \frac{(n+1)^{n}(n+1)n!^{2}}{(n+1)^{2}n!^{2}n^{n}} = \lim\limits_{n \to \infty} \frac{(n+1)^{n}}{(n+1)n^{n}} = 0 < 1 - $ ряд сходится

### 2. Исследовать ряд на сходимость, используя радикальный признак Коши:
### $ \sum_{n=1}^{\infty} \frac{n}{2^n}
$

### $\lim\limits_{n \to \infty} \sqrt[n]{\frac{n}{2^{n}}} = \frac{1}{2} < 1 \Rightarrow $ ряд сходится

### 3. Исследовать ряд на сходимость, используя признак Лейбница:
### $ \sum_{n=1}^{\infty} \frac{(-1)^{n}}{n+ln(n)}
$

### $ |-1| + |\frac{1}{2+ln2}| + |\frac{-1}{3+ln3}| + ... \\
1 > |\frac{1}{2+ln2}| > |\frac{-1}{3+ln3}| > ... - \ условие \ выполняется \\
\lim\limits_{n \to \infty} |\frac{(-1)^{n}}{n + ln(n)}| = \lim\limits_{n \to \infty} |\frac{1}{n + ln(n)}| = 0 - \  условие \ выполняется \\
Для \ сравнения \ берем \ гармонический \ ряд \sum_{n=1}^{\infty} \frac{1}{n} \\
\lim\limits_{n \to \infty} \frac{\frac{1}{n}}{\frac{1}{n+ln(n)}} = \lim\limits_{n \to \infty} \frac{{n+ln(n)}}{n} = 1 \Rightarrow ряд \ расходится$
### Ответ: Ряд сходится условно.
 

### 4. Исследовать ряд на сходимость, используя признак Раабе:
### $ \sum_{n=1}^{\infty} \frac{3^{n}}{2^{n}}
$
